```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import TanimotoSimilarity
```

```python
print(patient["SMILES (drugA)"].isnull().sum())
print(patient["SMILES (drugB)"].isnull().sum())
```

```python
# Fill missing values with an empty string
#df['smiles1'] = df['smiles1'].fillna('')

# Alternatively, drop rows with missing values
patient = patient.dropna(subset=['SMILES (drugA)', 'SMILES (drugB)'])
```

```python
patient['molA'] = patient['SMILES (drugA)'].apply(Chem.MolFromSmiles)
patient['molB'] = patient['SMILES (drugB)'].apply(Chem.MolFromSmiles)
patient.head()
```

```python
from rdkit.Chem import rdFingerprintGenerator
# Create a Morgan fingerprint generator
fpgen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=1024)

patient['fpA'] = patient['molA'].apply(fpgen.GetFingerprint)
patient['fpB'] = patient['molB'].apply(fpgen.GetFingerprint)
patient.head()
```

```python
# Calculate Tanimoto coefficient
patient['tanimoto'] = patient.apply(lambda row: TanimotoSimilarity(row['fpA'], row['fpB']), axis=1)
```

```python
# Print the results
print(patient[['SMILES (drugA)', 'SMILES (drugB)', 'tanimoto']])
```

## Adding Carrier, Enzyme, Transport and Target Vectors

```python
carrier = pd.read_csv("carrier/all.csv")
enzyme = pd.read_csv("enzyme/all.csv")
target = pd.read_csv("target/all.csv")
transporter = pd.read_csv("transporter/all.csv")
```

```python
binary_features_dfs = [carrier, enzyme, target, transporter]
```

```python
def build_drugID_list(df):
    df['Drug IDs'] = df['Drug IDs'].apply(lambda x: x.split('; '))
    
for df in binary_features_dfs:
    build_drugID_list(df)
```

```python
binary_features_dfs[0]["Drug IDs"]
```

```python
def create_binary_vector(drug_id, carrier_lists):
    return [1 if drug_id in carrier_list else 0 for carrier_list in carrier_lists]

patient["carrier_bvA"] = patient["DrugBank ID A"].apply(lambda x: create_binary_vector(x, carrier["Drug IDs"]))
patient["enzyme_bvA"] = patient["DrugBank ID A"].apply(lambda x: create_binary_vector(x, enzyme["Drug IDs"]))
patient["target_bvA"] = patient["DrugBank ID A"].apply(lambda x: create_binary_vector(x, target["Drug IDs"]))
patient["transporter_bvA"] = patient["DrugBank ID A"].apply(lambda x: create_binary_vector(x, transporter["Drug IDs"]))
patient["carrier_bvB"] = patient["DrugBank ID B"].apply(lambda x: create_binary_vector(x, carrier["Drug IDs"]))
patient["enzyme_bvB"] = patient["DrugBank ID B"].apply(lambda x: create_binary_vector(x, enzyme["Drug IDs"]))
patient["target_bvB"] = patient["DrugBank ID B"].apply(lambda x: create_binary_vector(x, target["Drug IDs"]))
patient["transporter_bvB"] = patient["DrugBank ID B"].apply(lambda x: create_binary_vector(x, transporter["Drug IDs"]))
```

```python
patient.columns
```

```python
import numpy as np

# Function to concatenate binary vectors
def concatenate_vectors_A(row):
    return np.concatenate([row["carrier_bvA"], row['enzyme_bvA'], row['target_bvA'], row['transporter_bvA']])

def concatenate_vectors_B(row):
    return np.concatenate([row['carrier_bvB'], row['enzyme_bvB'], row['target_bvB'], row['transporter_bvB']])

# Apply the function to each row to create a new column with combined vectors
patient['combined_vector_A'] = patient.apply(concatenate_vectors_A, axis=1)
patient['combined_vector_B'] = patient.apply(concatenate_vectors_B, axis=1)
```

### Adding Jaccard Similarity

```python
def jaccard_similarity(v1, v2):
    intersection = sum(1 for i, j in zip(v1, v2) if i == j == 1)
    union = sum(1 for i, j in zip(v1, v2) if i == 1 or j == 1)
    return intersection / union if union != 0 else 0

#patient['carrier_jsim'] = patient.apply(lambda row: jaccard_similarity(row['carrier_bvA'], row['carrier_bvB']), axis=1)
#patient['enzyme_jsim'] = patient.apply(lambda row: jaccard_similarity(row['enzyme_bvA'], row['enzyme_bvB']), axis=1)
#patient['target_jsim'] = patient.apply(lambda row: jaccard_similarity(row['target_bvA'], row['target_bvB']), axis=1)
#patient['transporter_jsim'] = patient.apply(lambda row: jaccard_similarity(row['transporter_bvA'], row['transporter_bvB']), axis=1)
patient['feature_jsim'] = patient.apply(lambda row: jaccard_similarity(row['combined_vector_A'], row['combined_vector_B']), axis=1)
```

### Adding Dice and Ochaiai Similarity

```python
def calculate_dice_and_ochiai(row):
    A = np.array(row['combined_vector_A'])
    B = np.array(row['combined_vector_B'])
    
    # Calculate intersection
    intersection = np.sum(A & B)
    
    # Calculate Dice similarity
    dice_similarity = (2 * intersection) / (np.sum(A) + np.sum(B))
    
    # Calculate Ochiai similarity
    ochiai_similarity = intersection / np.sqrt(np.sum(A) * np.sum(B))
    
    return pd.Series([dice_similarity, ochiai_similarity])

# Apply the function to each row
patient[['feature_dsim', 'feature_osim']] = patient.apply(calculate_dice_and_ochiai, axis=1)
```
