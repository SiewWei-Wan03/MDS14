```python
carrier = pd.read_csv("carrier/all.csv")
enzyme = pd.read_csv("enzyme/all.csv")
target = pd.read_csv("target/all.csv")
transporter = pd.read_csv("transporter/all.csv")
```

```python
def create_binary_vector(drug_id, carrier_lists):
    return [1 if drug_id in carrier_list else 0 for carrier_list in carrier_lists]

final_SMILES["carrier_bvA"] = final_SMILES["DrugBank ID A"].apply(lambda x: create_binary_vector(x, carrier["Drug IDs"]))
final_SMILES["enzyme_bvA"] = final_SMILES["DrugBank ID A"].apply(lambda x: create_binary_vector(x, enzyme["Drug IDs"]))
final_SMILES["target_bvA"] = final_SMILES["DrugBank ID A"].apply(lambda x: create_binary_vector(x, target["Drug IDs"]))
final_SMILES["transporter_bvA"] = final_SMILES["DrugBank ID A"].apply(lambda x: create_binary_vector(x, transporter["Drug IDs"]))
final_SMILES["carrier_bvB"] = final_SMILES["DrugBank ID B"].apply(lambda x: create_binary_vector(x, carrier["Drug IDs"]))
final_SMILES["enzyme_bvB"] = final_SMILES["DrugBank ID B"].apply(lambda x: create_binary_vector(x, enzyme["Drug IDs"]))
final_SMILES["target_bvB"] = final_SMILES["DrugBank ID B"].apply(lambda x: create_binary_vector(x, target["Drug IDs"]))
final_SMILES["transporter_bvB"] = final_SMILES["DrugBank ID B"].apply(lambda x: create_binary_vector(x, transporter["Drug IDs"]))
```

```python
import numpy as np

# Function to concatenate binary vectors
def concatenate_vectors_A(row):
    return np.concatenate([row["carrier_bvA"], row['enzyme_bvA'], row['target_bvA'], row['transporter_bvA']])

def concatenate_vectors_B(row):
    return np.concatenate([row['carrier_bvB'], row['enzyme_bvB'], row['target_bvB'], row['transporter_bvB']])

# Apply the function to each row to create a new column with combined vectors
final_SMILES['combined_vector_A'] = final_SMILES.apply(concatenate_vectors_A, axis=1)
final_SMILES['combined_vector_B'] = final_SMILES.apply(concatenate_vectors_B, axis=1)
```

```python
def jaccard_similarity(v1, v2):
    intersection = sum(1 for i, j in zip(v1, v2) if i == j == 1)
    union = sum(1 for i, j in zip(v1, v2) if i == 1 or j == 1)
    return intersection / union if union != 0 else 0

final_SMILES['feature_jsim'] = final_SMILES.apply(lambda row: jaccard_similarity(row['combined_vector_A'], row['combined_vector_B']), axis=1)
```

```python
final_SMILES[["tanimoto", "feature_jsim"]]
```

```python
def calculate_dice_and_ochiai(row):
    A = np.array(row['combined_vector_A'])
    B = np.array(row['combined_vector_B'])
    
    # Calculate intersection
    intersection = np.sum(A & B)
    
    # Calculate Dice similarity
    dice_similarity = (2 * intersection) / (np.sum(A) + np.sum(B))
    
    # Calculate Ochiai similarity
    ochiai_similarity = intersection / np.sqrt(np.sum(A) * np.sum(B))
    
    return pd.Series([dice_similarity, ochiai_similarity])

# Apply the function to each row
final_SMILES[['feature_dsim', 'feature_osim']] = final_SMILES.apply(calculate_dice_and_ochiai, axis=1)
```
